```python
# 2.Create a scatter plot to represent the data. What can you conclude from the scatterplot? Make sure you have the appropriate
# labels for the x and y axis.

sales = ["$215", "$325", "$185", "$332", "$406", "$522", "$412", "$614", "$544", "$421"]

# Remove the dollar sign from each element
sales = [sale[1:] for sale in sales]

# Convert the strings to integers
sales = [int(sale) for sale in sales]

temperature_with_degree = ["57.56", "61.52", "53.42", "59.36", "65.3", "71.78", "66.92", "77.18", "74.12", "64.58"]

# Remove the degree symbol from each element
temperature_without_degree = [temp[:-1] for temp in temperature_with_degree]

# Convert the strings to floating-point numbers
temperature_without_degree_float = [float(temp) for temp in temperature_without_degree]

print(temperature_without_degree_float)
# Create scatter plot
plt.scatter(sales,temperature_without_degree, color='skyblue', edgecolor='black')
plt.title('Scatter Plot of Sales vs Revenue')
plt.xlabel('Number of Sales')
plt.ylabel('Daily Revenue ($)')
plt.colorbar(label='Noon Temperature (C)')
plt.grid(True)
plt.show()
```

```python
# 3.1. Create a line graph that displays the total profit across all months. Make sure you have the appropriate labels for the x and y axis.

df = pd.read_csv("C:/Users/micha/PycharmProjects/Monthly Ice Cream Shop Sales by Flavor.csv")
# Create line graph for total profit across all months
df.head()
```

```python
plt.plot(df['Month'], df['Total Profit'], marker='o', color='skyblue', linestyle='-')
plt.title('Total Profit Across All Months')
plt.xlabel('Month')
plt.ylabel('Total Profit ($)')
plt.xticks(rotation=45)
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
# 3. 2. Create a bar chart that displays the total number of units sold per month. Make sure you have the appropriate labels for the x and y axis.
plt.bar(df['Month'], df['Total Units'], color='skyblue')
plt.title('Total Number of Units Sold Per Month')
plt.xlabel('Month')
plt.ylabel('Total Units Sold')
plt.xticks(rotation=45)
plt.grid(True)
plt.tight_layout()
plt.show()
```
